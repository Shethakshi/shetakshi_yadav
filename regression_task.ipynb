are not the major aspects one will consider while buying a house, which can also be the reason most of the data is missing.


WE CAN DROP THE VARIABLES/FEATURES WHICH HAVE HIGH PERCENTAGE OF MISSING VALUES, BECAUSE IF THERE ARE MORE THAN 50% OF 
MISSING VALUES THEN WE MIGHT AS WELL DROP THE FEATURE.

```python
# Calculate the missing value percentage for each column
missing_values_percentage = (ames_train.isnull().sum() / len(ames_train)) * 100
missing_values_percentage_test = (ames_test.isnull().sum() / len(ames_test)) * 100

# Identify columns with more than 15% missing values
columns_to_drop = missing_values_percentage[missing_values_percentage > 15].index
columns_to_drop_test = missing_values_percentage_test[missing_values_percentage_test > 15].index

# Drop the identified columns from the DataFrame
ames_train_dropped = ames_train.drop(columns=columns_to_drop)
ames_test_dropped = ames_test.drop(columns=columns_to_drop_test)

# Display the remaining columns
print("Remaining columns in train dataset:", ames_train_dropped.columns)
print("Remaining columns in test dataset:", ames_test_dropped.columns)



# Display the updated number of columns
print("Remaining columns in train dataset:", len(ames_train_dropped.columns))
print("Remaining columns in test dataset:", len(ames_test_dropped.columns))

```

#AS OUR MAIN FOCUS IS TARGET VARIABLE, WE WILL DROP THE FEATURES WHICH HAVE LESS THAN 0.3 CORRELATION WITH TARGET VARIABLE.
NOTE: I HAVE REFERENCED SNS(SEABORN) DOCUMENTATION FOR FEW CODE SNIPPETS AND HYPERPARAMETER TUNING BELOW.

```python
# Setting a correlation threshold of 30%
correlation_threshold = 0.3

# Calculating the correlation matrix
correlation_matrix = ames_train.corr()

# getting the correlation values with the target variable
correlation_with_saleprice = correlation_matrix["SalePrice"].abs()

# Identifying the features with correlation less than the threshold
less_correlated_features = correlation_with_saleprice[correlation_with_saleprice < correlation_threshold].index

# Dropping the less correlated features from the train dataset
ames_train_filtered = ames_train.drop(less_correlated_features, axis=1)

# Dropping the same features from the test dataset
ames_test_filtered = ames_test.drop(less_correlated_features, axis=1)

# Getting the count of remaining columns
column_count = ames_train_filtered.shape[1]
column_count_test= ames_test_filtered.shape[1]


# Display the count
print("Number of remaining columns:", column_count)
print("Number of remaining columns:", column_count_test)


# Displaying the remaining columns
print(ames_train_filtered.columns)
print(ames_test_filtered.columns)


# Updating the original train and test datasets with the filtered columns
ames_train = ames_train_filtered
ames_test = ames_test_filtered
```

INSIGHT: All the columns which are not much correlated with the target variable are dropped, but there are more features
    which have not much importance or do not contribute significantly to our problem statement
>> Also, there exists multicollinearity between few variables, in such cases we can drop the either one.
